```python
import numpy as np #грузим датасет. расписать где взял и что убрал
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('PS_2026.06.25_09.39.08.csv', skiprows=20)
data.sample(5)
```

```python
filtered_data = data.dropna(subset=['pl_radj', 'pl_bmassj'])


# Построение графика с поменянными осями
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['pl_bmassj'], filtered_data['pl_radj'], alpha=0.5, s=10, color='darkblue')

# Логарифмическая шкала теперь применяется к оси X (Масса)
plt.xscale('log')
plt.yscale('log')

# Обновляем подписи и заголовок
plt.title('Зависимость радиуса от массы экзопланет')
plt.xlabel('Масса планеты (в массах Юпитера) - логарифмическая шкала')
plt.ylabel('Радиус планеты (в радиусах Юпитера)')

#plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

```python
x_model = np.logspace(-2, 0, 1000) # 100 точек от 10^-1 до 10^3
y_model = x_model ** (1/3) # Вычисляем радиус по формуле y = x^3
plt.plot(x_model, y_model, color='blue', linewidth=2, label='Теоретическая зависимость M ∝ R³')

# 2. Рисуем реальные данные поверх линии
plt.scatter(filtered_data['pl_bmassj'], filtered_data['pl_radj'], 
            alpha=0.6, s=15, color='orange', label='Реальные экзопланеты')

# Устанавливаем логарифмический масштаб для ОБЕИХ осей
plt.xscale('log')
plt.yscale('log')

# Ограничиваем оси, чтобы график был наглядным
plt.xlim(0.001, 15) # Диапазон по массе
plt.ylim(0.1, 2.5) # Диапазон по радиусу

# Подписи, заголовок и легенда
plt.title('Сравнение реальных данных с зависимостью M ∝ R³')
plt.xlabel('Масса планеты (в массах Юпитера)')
plt.ylabel('Радиус планеты (в радиусах Юпитера)')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.4)

plt.show()
```

```python
# Очистка данных (та же самая)
filtered_data = data.dropna(subset=['pl_radj', 'pl_bmassj'])

# Построение графика
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_data, x='pl_radj', y='pl_bmassj', alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.title('Зависимость радиуса и массы экзопланет')
plt.xlabel('Радиус планеты (в радиусах Юпитера)')
plt.ylabel('Масса планеты (в массах Юпитера)')
plt.show()
```

```python

planets_mass = [
    3e-3,       # Земля
    0.046,      # Нептун
    1.0,        # Юпитер
    8e-5        # Плутон
]

planets_radius = [
    0.101,
    0.357,
    1.0,
    0.029
]

labels = ['Земля', 'Нептун', 'Юпитер', 'Плутон']

# 3. ПОСТРОЕНИЕ ФИГУРЫ И ОСНОВНОГО ГРАФИКА
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['pl_bmassj'], filtered_data['pl_radj'],
            alpha=0.5, s=10, color='darkblue')  # Основной график

# 4. НАКОЛАДКА ПЛАНЕТ
plt.scatter(
    planets_mass, planets_radius,
    marker='*', s=200, edgecolors='black', facecolor='gold',
    zorder=10  # Поверх облака
)

for mass, radius, label in zip(planets_mass, planets_radius, labels):
    plt.annotate(label, xy=(mass, radius), xytext=(mass * 1.1, radius * 1.1),
                 arrowprops={'arrowstyle': '-', 'lw': 0.5},
                 fontsize=8, ha='center')

# 5. Оформление
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.001, 50)  # 👉 Важно: увеличим левый порог, иначе Плутон уйдет за экран!
plt.ylim(0.07, 3)

plt.title('Экзопланеты vs Солнечная система')
plt.xlabel('Масса (в массах Юпитера)')
plt.ylabel('Радиус (в радиусах Юпитера)')

plt.show()
```

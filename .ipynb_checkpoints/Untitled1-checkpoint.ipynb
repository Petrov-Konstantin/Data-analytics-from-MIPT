```python
import numpy as np #грузим датасет. расписать где взял и что убрал
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('PS_2026.06.25_09.39.08.csv', skiprows=20)
data.sample(5)
```

```python
data[data.isnull().any(axis=1)]
```

```python
data_copy=data.copy()
data_copy.loc[data['pl_name'] == 'K2-24 b', 'disc_year'] = 2026
```

```python
cleaned_df = (
    data.groupby('pl_name')
    .agg({
        # Категории и статусы
        'sy_snum': 'first',
        'discoverymethod': 'first',
        'disc_year': 'last',  # Берем последний известный год
        'soltype': 'first',
        'pl_controv_flag': 'max',
        
        # Физические параметры
        'pl_radj': 'median',  # Медиану, чтобы убрать влияние выбросов
        'pl_bmassj': 'median'
    })
    .reset_index()
)
cleaned_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Удаляем строки с нулевыми или пропавшими значениями
clean_plot = cleaned_df[
    (cleaned_df['pl_radj'] > 0) &
    (cleaned_df['pl_bmassj'] > 0)
]



plt.figure(figsize=(10, 6))

# Основные точки
sns.scatterplot(
    data=clean_plot,
    x='pl_bmassj', y='pl_radj',
    hue='sy_snum', palette='viridis',  # Цвет по году открытия
    alpha=0.6, size=10
)

# Теоретическая зависимость M ~ R³
x_model = clean_plot['pl_bmassj'].quantile([0.001, 0.99]).values  # Возьмем крайние 1%
y_model = x_model ** (1/3)
plt.plot(x_model, y_model, '--k', lw=2, label='$M \\sim R^3$')

# Настройка лог-шкал
plt.xscale('log')
plt.yscale('log')

# Оси и сетка
plt.xlim(1e-3, 50)
plt.ylim(1e-2, 10)
plt.grid(which='both', ls=':', alpha=0.5)

# Оформление
plt.title('Зависимость массы от радиуса экзопланет (очищенный DF)')
plt.xlabel('Масса (в массах Юпитера)')
plt.ylabel('Радиус (в радиусах Юпитера)')
plt.legend(title='Год открытия');
```

```python
import numpy as np #грузим датасет. расписать где взял и что убрал
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('PS_2026.06.25_09.39.08.csv', skiprows=20)
data.sample(5)
```

```python
data.info()
```

```python
missing_rows = data[data.isnull().any(axis=1)]# ищем пропущенные значения
missing_rows
```

по названию ищем что не так с планетой

препринт статьи есть, давайте просто поставим 2026 год

```python
data_copy=data.copy()
data_copy.loc[data['pl_name'] == 'TOI-5789 c', 'disc_year'] = 2026.0
data_copy.info()
```

```python
pl_name_counts = data_copy['pl_name'].value_counts() # проблема: некоторые планеты упоминаются в разных статьях
pl_name_counts
```

```python
cleaned_df = (
    data.groupby('pl_name')
    .agg({
        # Категории и статусы
        'sy_snum': 'first',
        'discoverymethod': 'first',
        'disc_year': 'first',  # Берем последний известный год
        'soltype': 'first',
        'pl_controv_flag': 'max',
        
        # Физические параметры
        'pl_radj': 'median',  # Медиану, чтобы убрать влияние выбросов
        'pl_bmassj': 'median'
    })
    .reset_index()
)
cleaned_df.info()
```

```python
cleaned_df['sy_snum'].value_counts() #сколько звёзд в стстеме
```

```python
cleaned_df['discoverymethod'].value_counts() #какким образом открыли
```

```python
data_copy.describe() #средние значения планет. большие отклонения - разные типы+
```

```python
data_copy.groupby('sy_snum')['pl_bmassj'].describe()
```

```python
years = cleaned_df['disc_year'].dropna().round().astype(int)
ax = years.plot.hist(bins=range(1995, 2027), figsize=(10, 4))
ax.set_title('Количество открытых экзопланет по годам')
ax.set_ylabel('Количество планет')
ax.set_xlabel('Год открытия');# 2014 запуск Кеплера
```

```python
import seaborn as sns
# Очищаем от нулей и NaN
clean_plot = cleaned_df[
    (cleaned_df['pl_radj'] > 0) &
    (cleaned_df['pl_bmassj'] > 0)
]

# Вариант 1: Прописываем все (быстро, надежно)
'''custom_colors = {
    1: '#DC143C',  # Желтый
    2: '#1E90FF',  # Голубой
    3: '#1E90FF',  # Красный
    4: '#1E90FF'   # Серый (на всякий случай)
}'''

# Вариант 2: Оставляем только 1 (Python выбирает цвета сам)
# custom_colors = {1: '#FFA500'}

# Построение
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=clean_plot,
    x='pl_bmassj', y='pl_radj',
    alpha=0.6, s=10
)

# Оформление
plt.xscale('log'); plt.yscale('log')
plt.xlim(1e-3, 50)
plt.ylim(1e-2, 10)
plt.grid(True, which='both', ls=':')
#plt.legend(title='Количество звёзд в системе');
plt.title('Зависимость массы от радиуса экзопланет (очищенный DF)')
plt.xlabel('Масса (в массах Юпитера)')
plt.ylabel('Радиус (в радиусах Юпитера)')
```
